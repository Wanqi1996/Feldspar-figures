In this notebook, we illustrate the application of PyTorch to regress the aqueous solubility of different molecules using a neural network model.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap, PowerNorm

# 数据矩阵
data = np.array([
    [99.4774, 41.37234, 53.9099, 28.91931, 112.71178, 18.51334, 58.87328],
    [164.87385, 85.85245, 110.78328, 78.40909, 142.67639, 51.05183, 91.8328],
    [137.71217, 129.42198, 184.33256, 119.61579, 101.03645, 152.3961, 112.10709],
    [112.71285, 58.64779, 74.04719, 34.83175, 119.53552, 30.88595, 69.2836],
    [110.09792, 49.12517, 75.80228, 35.87823, 120.24581, 27.41517, 65.84218],
    [143.95067, 130.9454, 188.27086, 102.42912, 98.98416, 166.37803, 106.49248],
    [190.13089, 168.31042, 227.33423, 141.46027, 124.01732, 182.29476, 151.95825],
    [175.25632, 96.3453, 119.17045, 65.46626, 106.2915, 70.0413, 81.36079],
    [133.92249, 63.51084, 92.16396, 43.05599, 110.27855, 50.33144, 65.66411],
    [178.76589, 115.22549, 171.12689, 83.08885, 137.96754, 108.1803, 108.85314],
    [94.0213, 61.09438, 112.5873, 25.33959, 104.10734, 56.04272, 68.51254],
    [120.55425, 68.31705, 56.59825, 45.84226, 106.96769, 26.38274, 66.41939],
    [125.83197, 78.47201, 94.27918, 45.93396, 78.11054, 48.30402, 58.11459]
])

# 转置矩阵（交换x/y）
data = data.T

# 定义蓝到白的渐变色图
colors = [(0, 0, 0.8), (1, 1, 1)]  # 深蓝 -> 白
blue_white = LinearSegmentedColormap.from_list('blue_white', colors, N=256)

# 创建图像，比例2:1
plt.figure(figsize=(20, 5))

# 使用PowerNorm实现“指数分布”效果 (gamma < 1 增强低值对比)
#norm = PowerNorm(gamma=0.5, vmin=np.min(data), vmax=np.max(data))
norm = PowerNorm(gamma=0.5, vmin=np.min(data), vmax=150)

# 绘制热图
img = plt.imshow(data, cmap=blue_white, origin='lower', aspect='auto', interpolation='nearest', norm=norm)
plt.colorbar(label='数值大小（指数归一化）')

plt.title('二维数值分布图（蓝→白，指数分布，2:1 比例）')
plt.xlabel('X 方向')
plt.ylabel('Y 方向')

plt.show()
```
